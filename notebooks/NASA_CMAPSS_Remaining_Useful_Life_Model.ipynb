```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torch.optim as optim
import torch.utils.data as data
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
data_train = pd.read_csv('../data/processed/train_FD001.txt', sep=' ', header=None)
data_test = pd.read_csv('../data/processed/test_FD001.txt', sep=' ', header=None)
data_train.head()
```

```python
data_train.rename(columns={0: 'ID', 1: 'Cycles',2:"opsettings1",3:"opsettings2", 19:"Clustering"}, inplace=True)
for i in range(1,16):
    data_train.rename(columns={i+3: 'Sensor'+str(i)}, inplace=True)
data_train["RUL"] = data_train.groupby("ID")["Cycles"].transform(lambda x: x.max() - x)
data_test.rename(columns={0: 'ID', 1: 'Cycles',2:"opsettings1",3:"opsettings2", 19:"Clustering"}, inplace=True)
for i in range(1,16):
    data_test.rename(columns={i+3: 'Sensor'+str(i)}, inplace=True)
data_test["RUL"] = data_test.groupby("ID")["Cycles"].transform(lambda x: x.max() - x)
data_train.head()
```

```python
data_test.head()
```

```python
sensor_cols = [col for col in data_train.columns if 'Sensor' in col]
X_sensors_train = data_train[sensor_cols]
X_sensors_test = data_test[sensor_cols]
pca = PCA(n_components=8)
X_pca_train = pca.fit_transform(X_sensors_train)
X_pca_test = pca.fit_transform(X_sensors_test)
```

```python
X_pca_test
```

```python
df_train_pca = data_train.drop(columns=sensor_cols).reset_index(drop=True)
sensor_pca_cols_train = [f'pca_{i}' for i in range(X_pca_train.shape[1])]
df_train_pca[sensor_pca_cols_train] = pd.DataFrame(X_pca_train)

df_test_pca = data_test.drop(columns=sensor_cols).reset_index(drop=True)
sensor_pca_cols_test = [f'pca_{i}' for i in range(X_pca_test.shape[1])]
df_test_pca[sensor_pca_cols_test] = pd.DataFrame(X_pca_test)
```

```python
df_train_pca.head()
```

```python

window_size = 50
def create_Sets(data_train, window_size):
    X = []
    Y = []
    # Agrupamos por ID de motor
    for j, group in data_train.groupby("ID"):
        group = group.sort_values("Cycles")
        features = group.iloc[:, 6:].values  # columnas 2 hasta penúltima (sin ID, Cycles, ni RUL)
        rul_values = group["RUL"].values

        if len(group) >= window_size:
            for i in range(len(group) - window_size + 1):
                X.append(features[i:i + window_size])
                Y.append(rul_values[i + window_size - 1])
        else:
            print(f"Warning: Unit {j} has less than {window_size} cycles")
    X = np.array(X)
    Y = np.array(Y)
    return torch.tensor(X, dtype=torch.float32), torch.tensor(Y[:, None], dtype=torch.float32)
X_train, Y_train = create_Sets(df_train_pca, window_size)
X_test, Y_test = create_Sets(df_test_pca, window_size)
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size=0.2)
```

```python
class Predictions(nn.Module):
    def __init__(self):
        super().__init__()
        self.lstm = nn.LSTM(input_size=8, hidden_size=50, num_layers=3, batch_first=True)
        self.dropout = nn.Dropout(0.2)  # 20% de desconexión
        self.linear = nn.Linear(50, 1)
        self.relu = nn.ReLU()
    def forward(self, x):
        x, _ = self.lstm(x)
        # extract only the last time step
        x = x[:, -1, :]
        x = self.dropout(x)
        x = self.linear(x)
        x = self.relu(x)
        return x
```

```python

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Usando dispositivo:", device)

model = Predictions().to(device)
optimizer = torch.optim.AdamW(model.parameters())
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', patience=3, factor=0.5, verbose=True)
loss_fn = nn.MSELoss()

X_train = X_train.to(device)
Y_train = Y_train.to(device)
X_test = X_test.to(device)
Y_test = Y_test.to(device)
X_val = X_val.to(device)
Y_val = Y_val.to(device)

loader = data.DataLoader(data.TensorDataset(X_train, Y_train), shuffle=True, batch_size=20)

n_epochs = 40
for epoch in range(n_epochs):
    model.train()
    for X_batch, y_batch in loader:
        X_batch = X_batch.to(device)
        y_batch = y_batch.to(device)

        y_pred = model(X_batch)
        loss = loss_fn(y_pred, y_batch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    model.eval()
    with torch.no_grad():
        y_pred = model(X_train)
        train_rmse = torch.sqrt(loss_fn(y_pred, Y_train)).item()
        y_pred = model(X_val)
        val_rmse = torch.sqrt(loss_fn(y_pred, Y_val)).item()
    print("Epoch %d: train RMSE %.4f, val RMSE %.4f" % (epoch, train_rmse, val_rmse))
```

```python
t = len(Y_test)
model.eval()
with torch.no_grad():
    y_pred = model(X_test).cpu().numpy()
    plt.plot(np.arange(t), Y_test.cpu().numpy(), 'b', label='True Value')
    plt.plot(np.arange(t), y_pred, 'r', label='Prediction')
    plt.legend(loc='upper left')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

y_pred = model(X_test).cpu().detach().numpy().flatten()
y_true = Y_test.cpu().detach().numpy().flatten()

rmse =np.sqrt( mean_squared_error(y_true, y_pred))
mae = mean_absolute_error(y_true, y_pred)

print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
```
